```python
import numpy as np

X = []
Y = []

t = np.arange(0, 1, 0.01)
for i in range(0, 200):

    message_omega1 = np.random.uniform(1, 10) * 2 * np.pi
    message_omega2 = np.random.uniform(1, 10) * 2 * np.pi
    message_omega3 = np.random.uniform(10, 20) * 2 * np.pi
    message_omega4 = np.random.uniform(10, 20) * 2 * np.pi

    Carrier_wave_amplitude = 1.5

    m1 = np.sin(message_omega1*t) + np.sin(message_omega2*t)
    m2 = np.sin(message_omega3*t) + np.sin(message_omega4*t)


    X.append(m1.T)
    Y.append(np.array([0*t, 1*t]).T)
    X.append(m2.T)
    Y.append(np.array([1*t, 0*t]).T)


X = np.expand_dims(np.stack(X), axis=-1)
Y = np.stack(Y)

print(X.shape, Y.shape)

np.save('/X.npy', X)
np.save('Y.npy', Y)

print('X and Y saved as npy files')
np.save('X.npy', X)
np.save('Y.npy', Y)
```

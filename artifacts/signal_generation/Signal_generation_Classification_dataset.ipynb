```python
# Signal Generation Task
#######################################################################################
import numpy as np
import scipy.signal
import tensorflow as tf
from matplotlib import pyplot as plt
import random

# Step-size (1/fs)
t_step = 0.001
duration = 1
# Timescale (0, 1) with stepsize = 1/fs
t = np.arange(0, duration, t_step)
f = np.fft.rfftfreq(t.shape[0], t_step)

# Input frequency range (fmin, fmax) with freq_stepsize and amplitude range (0, amax) with amp_stepsize
freq_stepsize = 1
fmin = 0.1
fmax = 10
freqs = np.arange(fmin, fmax, freq_stepsize)
amp_stepsize = 0.1
amin = -3
amax = 3
amps = np.arange(amin, amax, amp_stepsize)

# Number of signals to be generated
dataset_size = 500

# Each input signal has freqcompnum frequency components
freqcompnum = 5

# Setting up all the arrays with right dimensions
ip_sigs = np.zeros((dataset_size, len(t)))
op_sigs = np.zeros((dataset_size, len(t), 2))
ip_fft = np.zeros((dataset_size, int(1 + (len(t)/2))))
op_fft = np.zeros((dataset_size, int(1 + (len(t)/2))))
op2_fft = np.zeros((dataset_size, int(1 + (len(t)/2))))

frequencies = np.zeros((dataset_size, freqcompnum))
amplitudes = np.zeros((dataset_size, freqcompnum))

# For generating the input signals
for sig in range(dataset_size):
    for i in range(freqcompnum):
        freq = random.choice(freqs)
        frequencies[sig][i] = freq
        amp = random.choice(amps)
        amplitudes[sig][i] = amp
        time_sample = 0
        for time in t:
            ip_sigs[sig][time_sample] += amp*np.cos(2*np.pi*freq*time)
            time_sample += 1

# Adding white noise (Different for all signals; 0 mean, 0.15 std)
for sig in range(dataset_size):
    ip_sigs[sig] += np.random.normal(0, 0.15, size= len(t))

for sig in range(dataset_size):
    op_sigs[sig] = np.array([1*t, 0*t]).T


# Input frequency range (fmin, fmax) with freq_stepsize and amplitude range (0, amax) with amp_stepsize
freq_stepsize = 1
fmin = 10.1
fmax = 20
freqs = np.arange(fmin, fmax, freq_stepsize)
amp_stepsize = 0.1
amin = -3
amax = 3
amps = np.arange(amin, amax, amp_stepsize)

# Number of signals to be generated
dataset_size = 500

# Each input signal has freqcompnum frequency components
freqcompnum = 5

# Setting up all the arrays with right dimensions
ip_sigs_2 = np.zeros((dataset_size, len(t)))
op_sigs_2 = np.zeros((dataset_size, len(t), 2))
ip_fft = np.zeros((dataset_size, int(1 + (len(t)/2))))
op_fft = np.zeros((dataset_size, int(1 + (len(t)/2))))
op2_fft = np.zeros((dataset_size, int(1 + (len(t)/2))))

frequencies = np.zeros((dataset_size, freqcompnum))
amplitudes = np.zeros((dataset_size, freqcompnum))

# For generating the input signals
for sig in range(dataset_size):
    for i in range(freqcompnum):
        freq = random.choice(freqs)
        frequencies[sig][i] = freq
        amp = random.choice(amps)
        amplitudes[sig][i] = amp
        time_sample = 0
        for time in t:
            ip_sigs_2[sig][time_sample] += amp*np.cos(2*np.pi*freq*time)
            time_sample += 1

# Adding white noise (Different for all signals; 0 mean, 0.15 std)
for sig in range(dataset_size):
    ip_sigs_2[sig] += np.random.normal(0, 0.15, size= len(t))

for sig in range(dataset_size):
    op_sigs_2[sig] = np.array([0*t, 1*t]).T
```
